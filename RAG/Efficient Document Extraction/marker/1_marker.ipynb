### Extract Markdown from PDFs

```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

converter = PdfConverter(
    artifact_dict=create_model_dict(),
)
rendered = converter("/home/qb-admin/Desktop/ap/budget_speech_2025.pdf")
text = text_from_rendered(rendered)
```

### Extract Markdown and images from PDF

```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered
import os

# Define the path to your PDF and the output directory
pdf_path = r"/home/qb-admin/Desktop/ap/budget_speech_2025.pdf"
output_dir = r"/home/qb-admin/Desktop/ap/Marker"

# Ensure the output directory exist
os.makedirs(output_dir, exist_ok=True)

# Initialize the PdfConverter with the model dictionary
converter = PdfConverter(artifact_dict=create_model_dict())

# Convert the PDF to a rendered object
rendered = converter(pdf_path)

# Extract text and images from the rendered object
text, _, images = text_from_rendered(rendered)

# Save the extracted Markdown text
markdown_path = os.path.join(output_dir, "output.md")
with open(markdown_path, "w", encoding="utf-8") as md_file:
    md_file.write(text)

# Save the extracted images
# for idx, image in enumerate(images):
#     image_path = os.path.join(output_dir, f"image_{idx + 1}.png")
#     image.save(image_path)

print(f"Markdown and images have been saved to {output_dir}")
```
